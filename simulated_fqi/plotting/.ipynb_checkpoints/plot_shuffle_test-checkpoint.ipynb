```python
dataset = []
data = []
for sample in results['fg']:
    dataset.append("Foreground")
    data.append(sample)
for sample in results['bg']:
    dataset.append("Background")
    data.append(sample)
data_dict = {"Dataset": dataset, "Performance": data}
df = pd.DataFrame.from_dict(data_dict)
df
```

```python
plt.figure(figsize=(7, 5))
plt.title("Structureless test")
sns.swarmplot(x="Dataset", y="Performance", data=df)
plt.legend()
plt.xlabel("Steps Survived")
plt.tight_layout()
#plt.savefig("../plots/shuffle_test.png")
plt.show()
```
