```python
import seaborn as sns
import matplotlib.pyplot as plt
import json
import scipy
import numpy as np
import pandas as pd
import matplotlib
from plottify import autosize
```

```python
# font = {"size": 15}
matplotlib.rc("font", **font)
# matplotlib.rcParams["text.usetex"] = True


with open("../shuffle_test.json") as f:
    results = json.load(f)
```

```python
dataset = []
data = []
for sample in results['fg']:
    dataset.append("Foreground")
    data.append(sample)
for sample in results['bg']:
    dataset.append("Background")
    data.append(sample)
data_dict = {"Dataset": dataset, "Performance": data}
df = pd.DataFrame.from_dict(data_dict)
```

```python
plt.figure(figsize=(7, 5))
plt.title("Structureless test")
sns.violinplot(x="Dataset", y="Performance", data=df, palette='muted')
plt.xlabel("Dataset")
plt.tight_layout()
plt.ylabel("Steps Survived")
plt.savefig("../plots/shuffle_test_violinplot.png")
# autosize()
plt.show()
```
